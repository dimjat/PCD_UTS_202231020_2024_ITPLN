```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

img = cv.imread('noisy.jfif')

img.shape

img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
res = cv.resize(img,None,fx=2,fy=2,
                interpolation = cv.INTER_CUBIC)
##atau
height, width = img.shape[:2]
res = cv.resize(img,(2*width, 2*height),
                interpolation = cv.INTER_CUBIC)

fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].imshow(img)
axs[1].imshow(res)
```

```python
img = cv.imread('noisy.jfif',0)
rows,cols =img.shape
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

M = npfloat32([[1,0,10],[0,1,10]])
img_translasi= cv.warpAffine(img,M,(cols,rows))

fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].imshow(img)
axs[1].imshow(img_translasi)
```

```python
img = cv.imread('dog.jfif',0)

img.shape
```

```python
rows,cols =img.shape
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
img_rotated= cv.warpAffine(img,M,(cols,rows))

fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].imshow(img)
axs[1].imshow(img_rotated)
```
